# SMS Data Exploration

This notebook loads the public SMS Spam Collection dataset from Hugging Face: `ucirvine/sms_spam`.

The source dataset labels messages as `ham` or `spam`. In this project, we map:
- `ham` → `benign`
- `spam` → `malicious`

For this first version, spam is used as a practical proxy for malicious SMS. This gives us a reproducible starting dataset for
binary SMS classification, even though future versions may use more specific malicious-message datasets.

```python
from sms_classifier.data import load_sms_dataset, save_splits

df = load_sms_dataset()
df.head()
```

```python
df.shape
```

```python
df["label"].value_counts()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

ax = sns.countplot(data=df, x="label", hue="label", palette="Set2", legend=False)
ax.set_title("SMS class distribution")
ax.set_xlabel("Class")
ax.set_ylabel("Message count")

plt.show()
```

```python
df_with_length = df.assign(message_length=df["message"].str.len())

df_with_length.groupby("label")["message_length"].describe()
```

```python
ax = sns.boxplot(
    data=df_with_length,
    x="label",
    y="message_length",
    hue="label",
    palette="Set2",
    legend=False,
)
ax.set_title("SMS message length by class")
ax.set_xlabel("Class")
ax.set_ylabel("Message length in characters")

plt.show()
```

## Train/validation/test split

The dataset is split into train, validation, and test sets using stratification so each split keeps approximately the same
benign/malicious class proportions.

```python
from sms_classifier.data import split_sms_dataset

train_df, validation_df, test_df = split_sms_dataset(df)
train_df.shape, validation_df.shape, test_df.shape
```

```python
split_proportions = {
    "train": train_df["label"].value_counts(normalize=True),
    "validation": validation_df["label"].value_counts(normalize=True),
    "test": test_df["label"].value_counts(normalize=True),
}
split_proportions
```

```python
save_splits(train_df, validation_df, test_df, output_dir="../data/processed")
```
